QUALITATIVE DATA laso known as categorical, if its values belong to a collection of known defined non-overlapping classes. Common examples include student letter grade (A, B, C, D or F), commercial bond rating (AAA, AAB, ...) and consumer clothing shoe sizes (1, 2, 3, ...).

```r
library(MASS)
painters
```

```r
painters$School
```

The FREQUENCY DISTRIBUTION of a data variable is a summary of the 
data occurrence in a collection of non-overlapping categories.

```r
school = painters$School
school.freq =table(school)
school.freq
cbind(school.freq)
```

The RELATIVE FREQUENCY distribution of a data variable is a summary of the frequency proportion in a collection of non-overlapping categories.

```r
school.relfreq = school.freq /nrow(painters)
school.relfreq
```

```r
old = options(digits=1)
school.relfreq
cbind(school.relfreq)
```

A BAR GRAPH of a qualitative data sample consists of vertical parallel bars that shows the frequency distribution graphically.

```r
barplot(school.freq)
```

```r
colors = c("red", "yellow", "green", "violet", "orange", "blue", "pink", "cyan") 
barplot(school.freq,col=colors)         # apply the barplot function 
                # set the color palette
```

A PIE CHART of a qualitative data sample consists of pizza wedges that shows the frequency distribution graphically.

```r
pie(school.freq)
```

```r
pie(school.freq, col=colors)
```
